```python
import requests
from bs4 import BeautifulSoup
import re

# 取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/34a7ae2f1982d008c5da93ef37496ae519666728'

# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)
# HTMLを解析する
soup = BeautifulSoup(response.text, 'html.parser')

article_body = soup.select_one('#uamods>div>div')

paragraphs = article_body.find_all('p')

cleaned = []
for p in paragraphs:
    # p の中に a が入っている場合は削除
    for a in p.find_all('a'):
        a.decompose()

    cleaned.append(p.get_text(strip=True))

txt = "\n".join(cleaned)
print(txt)

```
