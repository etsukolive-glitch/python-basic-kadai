```python
import requests
from bs4 import BeautifulSoup

# 取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/7e5453dda173a4ef42c386c3a3778a383a3856b2'

# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)

# HTMLを解析する
soup = BeautifulSoup(response.text, 'html.parser')

article = soup.find("article")

body = soup.select_one('article')
text = body.text
text
```
